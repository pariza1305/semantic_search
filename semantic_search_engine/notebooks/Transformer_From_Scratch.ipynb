# Transformer Architecture From Scratch

This notebook demonstrates the core components of the Transformer architecture from the *Attention Is All You Need* (2017) paper.

## 1. Setup and Imports

```python
import sys
import os
import torch

# Add the src directory to the path so we can import our custom module
sys.path.append(os.path.abspath(".."))
from src.transformer import ScaledDotProductAttention, MultiHeadAttention, TransformerEncoder
```

## 2. Scaled Dot-Product Attention

The core of the Transformer. The formula is:

$$Attention(Q, K, V) = softmax(\frac{QK^T}{\sqrt{d_k}})V$$

Let's test our from-scratch implementation.

```python
d_k = 64
seq_len = 10
batch_size = 2
num_heads = 1

# Create dummy queries, keys, and values
q = torch.randn(batch_size, num_heads, seq_len, d_k)
k = torch.randn(batch_size, num_heads, seq_len, d_k)
v = torch.randn(batch_size, num_heads, seq_len, d_k)

attention_layer = ScaledDotProductAttention(d_k)
output, weights = attention_layer(q, k, v)

print(f"Input shape: {q.shape}")
print(f"Output shape: {output.shape}")
print(f"Attention weights shape: {weights.shape}")

# Verify that attention weights sum to 1.0 along the sequence dimension
print(f"\nSum of attention weights (should be ~1.0): {weights[0, 0, 0, :].sum().item():.4f}")
```

## 3. The Full Transformer Encoder

Now we put it all together: Multi-Head Attention, Position-wise Feed Forward, Positional Encoding, and Residual Connections.

```python
d_model = 384
vocab_size = 30000
max_seq_len = 512

# Initialize our custom Encoder
encoder = TransformerEncoder(
    vocab_size=vocab_size,
    d_model=d_model,
    num_heads=6,
    num_layers=6,
    d_ff=1536,
    max_seq_len=max_seq_len
)

# Create some dummy input token IDs (e.g., representing a sentence)
dummy_input_ids = torch.randint(0, vocab_size, (batch_size, seq_len))

print(f"Input tokens shape: {dummy_input_ids.shape}")

# Pass through the encoder
encoder_output = encoder(dummy_input_ids)

print(f"Encoder output shape: {encoder_output.shape}")
print("Notice the output shape is (batch_size, seq_len, d_model), perfectly matching the original paper!")
```
